### Importing Libraries

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Preprocessing

```python
#create a function to get the importanat columns, since we have multiplte datasets
def get_cols(data):
    cols = ['Date and time']
    row = ["Gear oil temperature (°C)","Front bearing temperature (°C)","Rear bearing temperature (°C)","Gear oil inlet temperature (°C)","Generator bearing front temperature (°C)","Generator bearing rear temperature (°C)","Rotor bearing temp (°C)","Stator temperature 1 (°C)"]
    for i in data:
        for j in row:
            if j in i:
                if "(°C)" in i:
                    cols.append(i)
    new_df = pd.DataFrame(columns=cols)
    for i in new_df:
        new_df[i] = data[i]
    return new_df
```

```python
def combining_data(path):
    comb_data = pd.DataFrame(columns=['Date and time',"Gear oil temperature (°C)","Front bearing temperature (°C)","Rear bearing temperature (°C)","Gear oil inlet temperature (°C)","Generator bearing front temperature (°C)","Generator bearing rear temperature (°C)","Rotor bearing temp (°C)","Stator temperature 1 (°C)"])

    for i in os.listdir(path):
        df = pd.read_csv(path+'/'+i)
        df1 = get_cols(df)
        comb_data = pd.concat([comb_data, df1])
    return comb_data
```

```python
train = combining_data('../../Datasets/2020')
train.dropna(inplace=True)
test = combining_data('../../Datasets/2021')
test.dropna(inplace=True)
```

### Data Visualization and Understanding

<b>Gear Oil Temperature (°C)</b>: This refers to the overall temperature of the oil circulating within the gearbox of the wind turbine. The gearbox is a crucial component responsible for transferring mechanical energy from the rotating blades (rotor) to the generator, which converts it into electrical energy.
<br><br>
<b>Front Bearing Temperature (°C)</b>: This refers to the temperature of the main front bearing in the wind turbine's drivetrain. Bearings reduce friction between rotating parts, and their temperature is critical. If it exceeds normal operating limits, it can indicate wear, poor lubrication, or excessive load on the turbine.
<br><br>
<b>Rear Bearing Temperature (°C)</b>: This is the temperature of the main rear bearing. Like the front bearing, its temperature is monitored to ensure that it is functioning within safe operating limits. Any abnormal rise in temperature could be a sign of a lubrication issue, misalignment, or excessive wear.
<br><br>
<b>Gear Oil Inlet Temperature (°C)</b>: This refers to the temperature of the oil entering the gearbox, which is used for lubrication and cooling of the gears inside. The gearbox is a crucial part of the turbine that transfers mechanical power from the rotor to the generator. Monitoring oil temperature is important for preventing overheating and ensuring that the oil maintains its lubricating properties.
<br><br>
<b>Generator Bearing Front Temperature (°C)</b>: This is the temperature of the bearing located at the front of the generator. The generator converts the mechanical energy from the rotor into electrical energy. Bearings are essential to allow the rotor to spin freely, and high temperatures can lead to reduced efficiency or bearing failure.
<br><br>
<b>Generator Bearing Rear Temperature (°C)</b>: Similar to the front generator bearing, this measures the temperature of the bearing at the rear of the generator. Both front and rear bearings support the generator shaft and need to be closely monitored to ensure smooth operation.
<br><br>
<b>Rotor Bearing Temperature (°C)</b>: The rotor bearing is one of the most critical parts of the wind turbine as it supports the entire rotor assembly, which includes the blades. Its temperature must remain within a safe range to avoid potential damage or failure, as high rotor bearing temperatures can affect the efficiency of energy generation and can lead to catastrophic failures.
<br><br>
<b>Stator Temperature 1 (°C)</b>: The stator is the stationary part of the generator that interacts with the rotor to produce electricity. Monitoring the temperature of the stator is important as excessive heat can cause insulation breakdown, reduce efficiency, and potentially damage the generator.
