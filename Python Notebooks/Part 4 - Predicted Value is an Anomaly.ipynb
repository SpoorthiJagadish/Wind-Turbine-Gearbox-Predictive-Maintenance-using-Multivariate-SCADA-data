```python
import os
import pandas as pd
import numpy as np
from keras.models import model_from_json
```

```python
def data_into_timeseries(df):
    for i in df.columns:
        data = pd.DataFrame(df[i])
        data.index = df.index
        data.rename(columns={i:'y'}, inplace=True)
        data['x1']= data['y'].shift()
        data['x2']= data['y'].shift(2)
        data['x3']= data['y'].shift(3)
        data['x4']= data['y'].shift(4)
        data['x5']= data['y'].shift(5)
        data['x6']= data['y'].shift(6)

        data = data.iloc[6:]
        
        data.to_csv("../Datasets/Timeseries Data/Test/" + i + ".csv")
        
        print("Saved", i + ".csv")
```

```python
def check_anomaly(dataset_path, threshold):
    threshold=threshold*2
    for i in ["Generator bearing front temperature (Â°C).csv", "Generator bearing rear temperature (Â°C).csv", "Rotor bearing temp (Â°C).csv"]:
        df = pd.read_csv(dataset_path+i).values

        file = open('../Model Files/LSTM Model/model.json', 'r')
        loaded  = file.read()
        file.close()
        model = model_from_json(loaded)
        model.load_weights("../Model Files/LSTM Model/" + i + ".h5")
        
        for j in  df:
            dateTime = j[0]
            y = j[1]
            x = list(j[2:])
            
            pred = model.predict([[x]])[0][0]
            
            if pred-y > threshold:
                print("Anomaly Detected at", dateTime, "Predicted Temp:", pred, "Actual Temp:", y)
```

```python
check_anomaly(dataset_path="../Datasets/Timeseries Data/Test/", threshold=1.353)
```
